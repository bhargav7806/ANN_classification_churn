```python
import pandas as pd 
from tensorflow.keras.models import load_model
import pickle 
import tensorflow as tf 
import numpy as np
```

```python
model = load_model('model.h5')
with open('label_encoder_geo.pkl' , 'rb') as file:
    oht_encoder_geo = pickle.load(file)
with open('label_encoder_gender.pkl' , 'rb') as file:
    label_encoder_gender = pickle.load(file)
```

```python
with open('scaler.pkl' , 'rb') as file:
    scaler = pickle.load(file)
```

```python
input_data = {
    'CreditScore' : 608 , 
    'Geography': 'Spain',
    'Gender' :'Female',
    'Age' : 41 ,
    'Tenure' : 1 ,
    'Balance' : 80837.06 ,
    'NumOfProducts' : 1, 
    'HasCrCard' : 0,
    'IsActiveMember' : 1 , 
    'EstimatedSalary' : 112542.88
}
```

```python
geo_encoded = oht_encoder_geo.transform([[input_data['Geography']]])
geo_encoded = pd.DataFrame(geo_encoded , columns = oht_encoder_geo.get_feature_names_out(['Geography']))
```

```python
geo_encoded
```

```python
input_data = pd.concat([pd.DataFrame([input_data]).reset_index(drop = True) , geo_encoded]  , axis = 1)
```

```python
input_data
```

```python
input_data['Gender'] = label_encoder_gender.transform([input_data['Gender']])
input_data
```

```python
input_data.drop(columns = 'Geography' ,axis = 1, inplace = True)
```

```python
input_data.shape
```

```python
input_data = scaler.transform(input_data)
```

```python
prediction = model.predict(input_data)
```

```python
prediction
```

```python
if prediction > 0.5:
    print('the customer is likely to churn.')

else:
    print('this customer is not likely to churn')
```

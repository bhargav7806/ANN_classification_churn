```python
import pandas as pd 
from sklearn.model_selection import train_test_split 
from sklearn.preprocessing import StandardScaler , LabelEncoder
import pickle
```

```python
df = pd.read_csv("Churn_Modelling.csv")
df.head()
```

```python
df.drop(columns=['RowNumber' , 'CustomerId' , 'Surname'] , inplace = True)
df.head()
```

```python
df.isna().sum()
```

```python
encoder = LabelEncoder()
df['Gender'] = encoder.fit_transform(df['Gender'])
```

```python
df
```

```python
df['Geography'].unique()
```

```python
from sklearn.preprocessing import OneHotEncoder
oht_encoder = OneHotEncoder(sparse_output= False)
geo_encoder = oht_encoder.fit_transform(df[['Geography']])
```

```python
df
```

```python
oht_encoder.get_feature_names_out(['Geography'])
```

```python
geo = pd.DataFrame(geo_encoder, columns = ['Geography_France', 'Geography_Germany', 'Geography_Spain'])
```

```python
df = pd.concat([df , geo] , axis = 1)
```

```python
df
```

```python
import pickle 
with open("label_encoder_gender.pkl" , 'wb') as file:
    pickle.dump(encoder , file)

with open("label_encoder_geo.pkl" , 'wb') as file:
    pickle.dump(oht_encoder , file)

```

```python
x = df.drop(columns = ['Exited' , 'Geography'] , axis = 1)
y = df['Exited']
```

```python
x_train , x_test , y_train , y_test = train_test_split(x , y , test_size = 0.25 , random_state=42)
```

```python
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
```

```python
x_test = scaler.transform(x_test)
```

```python
x_train
```

```python
with open ('scaler.pkl' , 'wb') as file:
    pickle.dump(scaler , file)
```

### ANN

```python
import tensorflow as tf 
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense 
from tensorflow.keras.callbacks import EarlyStopping , TensorBoard
```

```python
x_train.shape[1]
```

```python
model = Sequential([
Dense(64 , activation='relu',input_shape =(x_train.shape[1],)),  ## first hidden layer connected with input layer  
Dense(32 , activation='relu'),  # second hidden layer
Dense(1 , activation='sigmoid')  # output layer
])
```

```python
model.summary()
```

```python
import tensorflow
opt = tensorflow.keras.optimizers.Adam(learning_rate = 0.01)
# opt2 = tensorflow.keras.optimizers.Adadelta(learning_rate = 0.01)
loss = tensorflow.keras.losses.BinaryCrossentropy()
```

```python
model.compile(optimizer  = "adam" , loss= "binary_crossentropy" , metrics = ['accuracy'])
```

```python
# Set up the tensorboard
import datetime
logs_dir = "logs/fit/" + datetime.datetime.now().strftime("%Y%m%d-%H%M%S")
```

```python
tensorflow_callback = TensorBoard(log_dir = logs_dir , histogram_freq=1)
```

```python
## set up early stopping 
early_stopping_callback = EarlyStopping(monitor='val_loss' , patience=5 , restore_best_weights=True)   ## if no improvement like no loss decreasing and nothing happening till 5 epoch stop process and take best weight untill now 
```

```python
history= model.fit(
    x_train , y_train , validation_data= (x_test , y_test ) , epochs = 100 , callbacks = [tensorflow_callback , early_stopping_callback]
)
```

```python
model.save("model.h5")
```
